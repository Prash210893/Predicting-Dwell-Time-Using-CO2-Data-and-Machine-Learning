```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, silhouette_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

def calculate_dwell_time(df):
    # Convert the Timestamp column to datetime format
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])
    
    # Initialize dwell time column
    df['Dwell_Time'] = 0
    
    # Calculate dwell time for each cluster
    df['Cluster_Change'] = (df['Cluster'] != df['Cluster'].shift()).astype(int)
    df['Cluster_Change_CumSum'] = df['Cluster_Change'].cumsum()
    
    dwell_times = df.groupby('Cluster_Change_CumSum')['Timestamp'].agg(['min', 'max'])
    dwell_times['Dwell_Time'] = (dwell_times['max'] - dwell_times['min']).dt.total_seconds() / 60
    
    # Map dwell time back to the original dataframe
    dwell_time_map = dwell_times['Dwell_Time'].to_dict()
    df['Dwell_Time'] = df['Cluster_Change_CumSum'].map(dwell_time_map)
    
    return df

def plot_elbow_method(data):
    inertia = []
    range_clusters = range(1, 11)  # Testing for 1 to 10 clusters
    for k in range_clusters:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
    
    plt.figure(figsize=(8, 6))
    plt.plot(range_clusters, inertia, marker='o')
    plt.title('Elbow Method for Optimal k')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.show()

def apply_kmeans_clustering(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data)
    return labels

def apply_dbscan(data):
    dbscan = DBSCAN(eps=0.5, min_samples=5)
    labels = dbscan.fit_predict(data)
    return labels

def apply_hierarchical_clustering(data):
    # Plot dendrogram
    Z = linkage(data, 'ward')
    plt.figure(figsize=(10, 7))
    dendrogram(Z)
    plt.title('Dendrogram for Hierarchical Clustering')
    plt.xlabel('Sample index')
    plt.ylabel('Distance')
    plt.show()
    
    # Apply Agglomerative Clustering
    agglomerative = AgglomerativeClustering(n_clusters=3)
    labels = agglomerative.fit_predict(data)
    return labels

def calculate_silhouette_score(data, labels):
    score = silhouette_score(data, labels)
    print(f'Silhouette Score: {score}')
    return score

# Load and process the multiple CO2 data files
file_paths = [
    r'C:\Users\joshi\Downloads\BillyCo2data.csv',
    r'C:\Users\joshi\Downloads\NealCo2data.csv',
    r'C:\Users\joshi\Downloads\LeightonCo2data.csv',
]
file_names = ['Billy', 'Neal', 'Leighton']
dataframes = []
for path in file_paths:
    df = pd.read_csv(path)
    try:
        split_columns = ['Timestamp', 'Entity Name', 'CO2 Level']
        df[split_columns] = df[df.columns[0]].str.split(';', expand=True)
        df['CO2 Level'] = pd.to_numeric(df['CO2 Level'])
        df['Timestamp'] = pd.to_datetime(df['Timestamp'])
        df.drop(columns=[df.columns[0]], inplace=True)
        dataframes.append(df)
    except Exception as e:
        print(f"Error processing file {path}: {e}")

# Define models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

# Process and save each person's data individually
for i, (df, name) in enumerate(zip(dataframes, file_names)):
    # Feature Engineering
    df['Rolling_avg_CO2'] = df['CO2 Level'].rolling(window=10, min_periods=1).mean()
    df['Lag_CO2'] = df['CO2 Level'].shift(1).fillna(method='bfill')
    df['Delta_CO2'] = df['CO2 Level'] - df['Lag_CO2']
    df['Hour_of_Day'] = df['Timestamp'].dt.hour
    df['Day_of_Week'] = df['Timestamp'].dt.dayofweek
    df['Rolling_std_CO2'] = df['CO2 Level'].rolling(window=10, min_periods=1).std().fillna(method='bfill')
    df['CO2_Diff_1'] = df['CO2 Level'].diff(1).fillna(method='bfill')
    df['CO2_Diff_2'] = df['CO2 Level'].diff(2).fillna(method='bfill')

    # Clustering to determine labels
    X = df[['CO2 Level', 'Rolling_avg_CO2', 'Lag_CO2', 'Delta_CO2', 'Hour_of_Day', 'Day_of_Week', 'Rolling_std_CO2', 'CO2_Diff_1', 'CO2_Diff_2']]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Elbow Method
    plot_elbow_method(X_scaled)
    
    # K-means Clustering
    labels_kmeans = apply_kmeans_clustering(X_scaled, n_clusters=3)
    df['Cluster'] = labels_kmeans
    label_mapping = {0: 'No one', 1: 'One person', 2: 'More than one person'}
    df['Occupancy_State'] = df['Cluster'].map(label_mapping)
    silhouette_kmeans = calculate_silhouette_score(X_scaled, labels_kmeans)
    
    # DBSCAN Clustering
    labels_dbscan = apply_dbscan(X_scaled)
    silhouette_dbscan = calculate_silhouette_score(X_scaled, labels_dbscan)
    
    # Hierarchical Clustering
    labels_hierarchical = apply_hierarchical_clustering(X_scaled)
    silhouette_hierarchical = calculate_silhouette_score(X_scaled, labels_hierarchical)
    
    # Save processed data
    processed_data_path = f'C:\\Users\\joshi\\Downloads\\{name}_processed_data.csv'
    df.to_csv(processed_data_path, index=False)
    print(f"{name}'s processed data has been saved to: {processed_data_path}")

    # Prepare data for model training
    features = df[['CO2 Level', 'Rolling_avg_CO2', 'Lag_CO2', 'Delta_CO2', 'Hour_of_Day', 'Day_of_Week', 'Rolling_std_CO2', 'CO2_Diff_1', 'CO2_Diff_2']]
    target = df['Occupancy_State']
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

    # Compare and evaluate models
    model_results = []
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True)
        model_results.append((model_name, accuracy, report))
        print(f"{model_name} Model Accuracy for {name}: {accuracy}")
        print(classification_report(y_test, y_pred))

    # Save model comparison results
    model_comparison_path = f'C:\\Users\\joshi\\Downloads\\{name}_model_comparison_results.csv'
    results_df = pd.DataFrame(model_results, columns=['Model', 'Accuracy', 'Report'])
    results_df.to_csv(model_comparison_path, index=False)
    print(f"{name}'s model comparison results have been saved to: {model_comparison_path}")

    # Use Random Forest for final predictions
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)
    df['Predicted_Occupancy_State'] = rf_model.predict(features)

    # Calculate dwell time and add it to the dataframe
    df = calculate_dwell_time(df)

    # Save the final predictions with dwell time
    final_predictions_path = f'C:\\Users\\joshi\\Downloads\\{name}_final_predictions_with_dwell.csv'
    df.to_csv(final_predictions_path, index=False)
    print(f"Random Forest Model final predictions with dwell time saved for {name} at: {final_predictions_path}")

```
